# Check outliers

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x='price')
plt.show()
```

```python
# Another method
# check outliers in price
Q1 = data['price'].quantile(0.05)
Q3 = data['price'].quantile(0.95)
print(Q1)
print(Q3)
# We call outlier if value is less than 5th percentile and greater than 95th percentile
# outlier = price(40-355)
IQR = Q3-Q1
print(IQR)
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
outliers = data[(data['price']<lower_bound) | (data['price']>upper_bound)]
outliers
```

# Handle Outliers

```python
# handling outliers for price
data.loc[data['price']>upper_bound, 'price'] = data['price'].median()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x='price')
plt.show()
```

# Data Wrangling

```python
# to change any column name
data.rename({'minimum_nights':'min_nights'},axis=1,inplace=True)
data.columns
```

```python
data.rename({'number_of_reviews':'no_of_reviews'},axis=1)
```

```python
data.head(2)
```

```python
# to drop columns
data.drop('longitude', axis=1)
```

```python
# to drop multiple columns
data.drop(['longitude', 'latitude'], axis=1)
```

```python
# to create new columns
data['Rating'] = data['number_of_reviews']/data['reviews_per_month']
data.head(5)
```

```python
# to convert last_review	into proper date format
data['last_review']=pd.to_datetime(data['last_review'],errors='coerce')  # already in the proper date format
```

```python
# to fetch years from last_review
data['year']= data['last_review'].dt.year
data.head(2)
```

```python
# to fetch month name from last_review
data['month']= data['last_review'].dt.month_name()
data.head(2)
```

```python
# apply---->>>>applies function to every row in a pandas dataframe
# converting neighbourhood into upper case
def upper_case(value):
  return value.upper()
data['neighbourhood']=data['neighbourhood'].apply(upper_case)
data.head(2)
```

```python
data['neighbourhood']=data['neighbourhood'].apply(lambda x: x.upper())
data.head(2)
```

```python
# rounded latitude to 2 decimal values
data['latitude']= data['latitude'].apply(lambda x: round(x,2))
data.head(2)
```

```python
# create new column by encoding -ve values as negative and +ve values as positve of longitude
data['longitude_value']= data['longitude'].apply(lambda x: 'positive' if x>0 else 'negative')
data.head(2)
```

# Conclusion:
This project provided valuable insights into the Airbnb landscape in NYC during 2019. Through data cleaning, exploratory data analysis, and visualization, we uncovered key trends and patterns that can benefit both hosts and travelers.

Key Findings:

1) Manhattan emerged as the most popular and expensive borough for Airbnb listings, with the highest booking prices and the largest share of available listings throughout the year.
2) Entire home/apt rentals commanded higher prices and received the highest ratings compared to private rooms and shared rooms.
3) Booking prices showed a general upward trend over the years, with 2019 recording the highest overall booking revenue.
4) Latitude and longitude did not exhibit a significant correlation with price or popularity, suggesting other factors play a more prominent role in determining these metrics.
5) While outliers were identified and handled, they provided further insight into the diverse nature of the NYC Airbnb market.

Recommendations:

1) Hosts in Manhattan can leverage the high demand and pricing potential by optimizing their listings for amenities and services that attract travelers.
2) Travelers seeking budget-friendly options should consider exploring listings outside Manhattan, particularly in Brooklyn and Queens, which offer more affordable accommodations.
3) Hosts listing entire home/apt rentals should focus on maintaining high standards to justify the premium prices and attract more bookings.
4) Both hosts and travelers can utilize insights from this analysis to make informed decisions based on pricing trends, availability, and accommodation preferences.
